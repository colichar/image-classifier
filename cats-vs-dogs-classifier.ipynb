# Cats vs Dogs Image Classifier

```python
import tensorflow_datasets as tfds
import tensorflow as tf

from tensorflow.keras import layers
from tensorflow.keras.models import Sequential

from tensorflow.keras.losses import SparseCategoricalCrossentropy
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint

import matplotlib.pyplot as plt
```

## A simple CNN

Downloading the image dataset from the [TensorFlow Datasets Catalog](https://www.tensorflow.org/datasets/catalog/cats_vs_dogs).

```python
(train_ds, val_ds, test_ds), metadata = tfds.load(
    'cats_vs_dogs',
    split=['train[:20%]', 'train[20%:25%]', 'train[25%:30%]'],
    with_info=True,
    as_supervised=True,
)
```

Printing an image from the dataset.

```python
get_label_name = metadata.features['label'].int2str

image, label = next(iter(train_ds))
_ = plt.imshow(image)
_ = plt.title(get_label_name(label))
```

Defining some helping functions
- Normalizing the values of pixels in images in range 0-1 with the function `_normalize_img`
- Preparing images for a sequential model with the function `configure_for_performance`

```python
batch_size=32
AUTOTUNE = tf.data.AUTOTUNE

def _normalize_img(img, label):
    img = tf.cast(img, tf.float32) / 255.
    img = tf.image.resize(img,[224,224])
    return (img, label)

def configure_for_performance(ds):
    ds = ds.cache()
    ds = ds.shuffle(buffer_size=1000)
    ds = ds.map(_normalize_img)
    ds = ds.batch(batch_size)
    ds = ds.prefetch(buffer_size=AUTOTUNE)
    return ds

train_ds = configure_for_performance(train_ds)
val_ds = configure_for_performance(val_ds)
```

A simple sequential model consisting from convolutional and max pooling layers which are standard for image processing problems. \
At the end of the model there is a dense layer with 2 neural cells which will be representative of the models confidence that the processed image is a cat or a dog. \
Commented out
- a few augmentation layers in the beginning which can be used to reduce the overfitting of the model, if present
- additional dense layers to help form a more complicated mapping from image -> dog or cat, if needed

```python
model = Sequential([
    #augmentation layers
#    layers.RandomFlip(mode='horizontal', input_shape=(224, 224, 3)),
#    layers.RandomRotation(0.2),
    
    #cnn layers
    layers.Conv2D(filters=32, kernel_size=(3, 3), activation='relu', padding='same', input_shape=(224, 224, 3)),
    layers.MaxPool2D(pool_size=(2, 2), strides=2),
    layers.Conv2D(filters=64, kernel_size=(3, 3), activation='relu', padding='same'),
    layers.MaxPool2D(pool_size=(2, 2), strides=2),
    layers.Conv2D(filters=128, kernel_size=(3, 3), activation='relu', padding='same'),
    layers.MaxPool2D(pool_size=(2, 2), strides=2),
    layers.Conv2D(filters=256, kernel_size=(3, 3), activation='relu', padding='same'), #, kernel_regularizer =tf.keras.regularizers.l2(l=0.01)),
    layers.MaxPool2D(pool_size=(2, 2), strides=2),
    layers.Conv2D(filters=512, kernel_size=(3, 3), activation='relu', padding='same'), #, kernel_regularizer =tf.keras.regularizers.l2(l=0.01)),
    layers.MaxPool2D(pool_size=(2, 2), strides=2),
    layers.Flatten(),
    
    #dense layers
#   Dense(units=256, activation='relu', kernel_regularizer =tf.keras.regularizers.l2(l=0.01)),
#   Dense(units=64, activation='relu'),
    layers.Dense(units=2, activation='softmax')
])

model.summary()
```

```python
model.compile(optimizer='adam',
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])
```

```python
epochs=10
history = model.fit(
  train_ds,
  validation_data=val_ds,
  epochs=epochs
)
```

```python
## Plots the loss and accuaracy of the model on train and validation set
# large gap between train & dev set accuracy means high variance/ overfitting
# small accuaracy of train set means high bias

def show_history(history):
    fig, ax = plt.subplots(1, 2, figsize=(15,5))
    ax[0].grid()
    ax[0].set_title('loss')
    ax[0].plot(history.epoch, history.history["loss"], label="Train loss")
    ax[0].plot(history.epoch, history.history["val_loss"], label="Validation loss")
    ax[1].grid()
    ax[1].set_title('acc')
    ax[1].plot(history.epoch, history.history["accuracy"], label="Train acc")
    ax[1].plot(history.epoch, history.history["val_accuracy"], label="Validation acc")
    ax[0].legend()
    ax[1].legend()
```

```python
show_history(history)
```

## Transfer Learning

Transfer learning is a method used mostly in the case when the available dataset is rather small. It uses a model which was trained on a similar dataset. Depending on "how" similar the datasets (or rather problem to be solved) are, we freeze some layers and their weights of the pre-trained model and do not change them during the training. Other layers and their weights will be trained on our new dataset. Also we are free to modify these remaining layers, completely remove or replace them.

First we download a signficant smaller dataset sample.

```python
(train_ds, val_ds, test_ds), metadata = tfds.load(
    'cats_vs_dogs',
    split=['train[:5%]', 'train[5%:6%]', 'train[6%:7%]'],
    with_info=True,
    as_supervised=True,
)
```

```python
get_label_name = metadata.features['label'].int2str

image, label = next(iter(train_ds))
_ = plt.imshow(image)
_ = plt.title(get_label_name(label))
```

Defining some helping functions

- Normalizing the values of pixels in images according to the `preprocess_input` function used for the `VGG16` model with the function `_normalize_img`
- Preparing images for a sequential model with the function `configure_for_performance`

```python
from tensorflow.keras.applications.vgg16 import preprocess_input

batch_size=32
AUTOTUNE = tf.data.AUTOTUNE

def _normalize_img(img, label):
    img = preprocess_input(tf.cast(img, tf.float32))
    img = tf.image.resize(img,[224,224])
    return (img, label)

def configure_for_performance(ds):
    ds = ds.cache()
    ds = ds.shuffle(buffer_size=1000)
    ds = ds.map(_normalize_img)
    ds = ds.batch(batch_size)
    ds = ds.prefetch(buffer_size=AUTOTUNE)
    return ds

train_ds = configure_for_performance(train_ds)
val_ds = configure_for_performance(val_ds)
test_ds = configure_for_performance(test_ds)
```

In this example of transfer learning we'll be using the VGG16 model which was trained on a subset of the ImageNet dataset. We freeze all but the last two layers with their weights. The second to last layer will be set as trainable and the last layer is going to be replaced with a 2 cell decision layer.

```python
from tensorflow.keras.applications.vgg16 import VGG16

# load vgg16 model
vgg_model = VGG16()

# create a tf sequential model and add all but the last layer from vgg16
model = Sequential()
for layer in vgg_model.layers[0:-1]:
    model.add(layer)

# set all but the last layer as not trainable
for layer in model.layers:
    layer.trainable = False
model.layers[-1].trainable=True

# add last decision layer
model.add(layers.Dense(2, activation='softmax'))

model.summary()
```

```python
# compile model
model.compile(
    optimizer=Adam(learning_rate=0.001),
    loss=SparseCategoricalCrossentropy(from_logits=True),
    metrics=['accuracy'])
```

We define two callbacks

- `early_stopping` will be called if a monitored value is not improving after a set number of epochs
- `model_checkpoint` will save the state of a model at a local location (here `'/tmp/checkpoint'`) when a best value for a monitored value is reached

```python
# add early stopping callback in case model is not improving anymore
early_stopping = EarlyStopping(
    monitor="val_loss",
    patience=2,
    verbose=2
)

# add checkpoint to retrieve best model weights
model_checkpoint = ModelCheckpoint(
    '/tmp/checkpoint',
    monitor='val_loss',
    mode='min',
    save_best_only=True,
    save_weights_only=True,
    verbose=1,
)
```

```python
# train the model
history = model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=5,
    callbacks=[early_stopping, model_checkpoint]
)
```

```python
# retrieve best weights for model
model.load_weights('/tmp/checkpoint')
```

```python
# evaluate the model
loss, acc = model.evaluate(
    x=test_ds,
    verbose=2
)
```

```python
show_history(history)
```

When comparing the results to the previous simple classifier, we notice that we achieved better accuracy with a significantly smaller dataset and number of epochs thanks to the method of transfer learning.\
It seems that the `val_los` is converging around `0.25` and not going to be lower. To further decrease `val_los` and increase `val_acc` we would need to get a larger dataset or do some data augmentation on the existing dataset.
